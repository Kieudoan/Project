# Tiền xử lý dữ liệu cho dữ liệu suy tim

```python
import pandas as pd
df=pd.read_csv('Nhom 8_Data4AssociationRule.csv')
# Giảm chiều dữ liệu
df= df[['Diabetes','Family History','Smoking','Obesity','Alcohol Consumption',
        'Previous Heart Problems','Medication Use','Heart Attack Risk']]
# Chỉ chọn các trường hợp có khả năng bị suy tim để phân tích
df= df.loc[df['Heart Attack Risk']==1].reset_index(drop=True)
df
```

# Tìm tập phổ biến và luật kết hợp

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
# Tìm tập phổ biến khi min_sup = 0.5 bằng thuật toán apriori
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
# Tìm luật kết hợp từ các tập phổ biến đã lọc được ở trên với giá trị min_conf = 0.8
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
# In ra các tập phổ biến
print("Frequent Itemsets:")
print(frequent_itemsets)
# In ra các luật tìm được
print("\nAssociation Rules:")
final_df = rules.loc[:, ['antecedents', 'consequents', 'confidence']]
print(final_df.to_string())
```
